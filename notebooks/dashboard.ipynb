A Ccontinuación se define una función para analizar la cantidad y presencia de valores faltantes:

```python
def analizar_na(df, umbral_alto=20):
    """
    Analiza valores NA en un DataFrame.
    
    Args:
        df: DataFrame a analizar
        mostrar_top: Número de columnas con más NA a mostrar
        umbral_alto: Porcentaje para considerar "alto" nivel de NA
    
    Returns:
        DataFrame con estadísticas de NA
    """

    na_count = df.isnull().sum()
    na_percent = (na_count / len(df)) * 100

    resultados = pd.DataFrame({
        'columna': na_count.index,
        'na_count': na_count.values,
        'na_percent': na_percent.values,
        'tipo_dato': df.dtypes.values
    })
    
    resultados = resultados.sort_values('na_percent', ascending=False).reset_index(drop=True)
    
    total_na = na_count.sum()
    total_celdas = df.shape[0] * df.shape[1]
    total_percent = (total_na / total_celdas) * 100
    
    print(" Resumen de NAs")
    print(f" Dimensiones del dataset: {df.shape[0]} filas y {df.shape[1]} columnas")
    print(f" resgistros totales: {total_celdas:,}")
    print(f" NA totales: {total_na:,}")
    print(f" Porcentaje total NA: {total_percent:.2f}%")
    
resultados_na = analizar_na(df_subset, umbral_alto=10)
```

### Organización de datos para dashboard

Para realizar un análisis correcto de en el dashboard, es necesaria una organización previa donde se clasifique cada registro según el año y las categorías a las que pertenezca.

```python
import pandas as pd

def transform_to_hierarchy(df):
    """
    Transforma el DataFrame de formato ancho a formato largo con estructura jerárquica
    """
    result_rows = []
    
    hierarchy_map = {
        # Out of pocket
        '     Out of pocket': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Out of pocket',
            'sub2': 'none',
            'sub3': 'none'
        },
        # Health Insurance - Private
        '          Private Health Insurance': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Health Insurance',
            'sub2': 'Private Health Insurance',
            'sub3': 'none'
        },
        # Health Insurance - Medicare
        '          Medicare': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Health Insurance',
            'sub2': 'Medicare',
            'sub3': 'none'
        },
        # Health Insurance - Medicaid
        '          Medicaid (Title XIX)': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Health Insurance',
            'sub2': 'Medicaid (Title XIX)',
            'sub3': 'none'
        },
        # Health Insurance - CHIP
        '          CHIP (Title XIX and Title XXI)': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Health Insurance',
            'sub2': 'CHIP (Title XIX and Title XXI)',
            'sub3': 'none'
        },
        # Defense & Veterans
        '         Department of Defense': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Health Insurance',
            'sub2': 'Department of Defense',
            'sub3': 'none'
        },
        '         Department of Veterans Affairs': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Health Insurance',
            'sub2': 'Department of Veterans Affairs',
            'sub3': 'none'
        },
        # Other Third Party
        '         Worksite Health Care': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'Worksite Health Care',
            'sub3': 'none'
        },
        '         Other Private Revenues': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'Other Private Revenues',
            'sub3': 'none'
        },
        '         Indian Health Services ': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'Indian Health Services',
            'sub3': 'none'
        },
        "         Workers' Compensation": {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': "Workers' Compensation",
            'sub3': 'none'
        },
        '         General Assistance': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'General Assistance',
            'sub3': 'none'
        },
        # Maternal/Child Health
        '         Maternal/Child Health': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'Maternal/Child Health',
            'sub3': 'none'
        },
        # Vocational Rehabilitation
        '         Vocational Rehabilitation': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'Vocational Rehabilitation',
            'sub3': 'none'
        },
        # Other Third Party - misc
        '         Other Federal Programs*': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'Other Federal Programs',
            'sub3': 'none'
        },
        '         SAMHSA': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'SAMHSA',
            'sub3': 'none'
        },
        '         Other State and Local Programs**': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'Other State and Local Programs',
            'sub3': 'none'
        },
        '         School Health': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Other Third Party Payers and Programs',
            'sub2': 'School Health',
            'sub3': 'none'
        },
        # Investment
        '         Research': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Investment',
            'sub2': 'Research',
            'sub3': 'none'
        },
        '         Structures & Equipment': {
            'cat': 'Total National Health Expenditures',
            'sub1': 'Investment',
            'sub2': 'Structures & Equipment',
            'sub3': 'none'
        }
    }
    
    federal_state_columns = [col for col in df.columns if 'Federal' in col or 'State and Local' in col]
    
    for _, row in df.iterrows():
        year = row['Year']
        
        for col in df.columns:
            if col == 'Year':
                continue
                
            if col in ['Total National Health Expenditures', 
                      '     Health Insurance',
                      '     Other Third Party Payers and Programs',
                      '     Public Health Activity',
                      '     Investment',
                      'Total CMS Programs (Medicaid, CHIP and Medicare)']:
                continue
            
            if col in hierarchy_map:
                value = row[col]
                h = hierarchy_map[col]
                
                result_rows.append({
                    'Category': h['cat'],
                    'Subcategory 1': h['sub1'],
                    'Subcategory 2': h['sub2'],
                    'Subcategory 3': h['sub3'],
                    'Value': value,
                    'Year': year
                })

    cols_list = df.columns.tolist()
    
    for idx, col in enumerate(cols_list):
        if col == '                          Federal  ' or col == '                          State and Local':
            # Buscar la categoría padre anterior
            parent_idx = idx - 1
            while parent_idx >= 0 and (cols_list[parent_idx].startswith('                          ')):
                parent_idx -= 1
            
            if parent_idx >= 0:
                parent_col = cols_list[parent_idx]
                sub3_value = 'Federal' if 'Federal' in col else 'State and Local'
                
                # Determinar subcategory 2 basado en el padre
                if 'Medicaid' in parent_col:
                    sub2 = 'Medicaid (Title XIX)'
                    sub1 = 'Health Insurance'
                elif 'CHIP' in parent_col:
                    sub2 = 'CHIP (Title XIX and Title XXI)'
                    sub1 = 'Health Insurance'
                elif 'Maternal' in parent_col:
                    sub2 = 'Maternal/Child Health'
                    sub1 = 'Other Third Party Payers and Programs'
                elif 'Vocational' in parent_col:
                    sub2 = 'Vocational Rehabilitation'
                    sub1 = 'Other Third Party Payers and Programs'
                elif 'Public Health' in parent_col:
                    sub2 = sub3_value
                    sub1 = 'Public Health Activity'
                    sub3_value = 'none'
                else:
                    continue
                
                for _, row in df.iterrows():
                    result_rows.append({
                        'Category': 'Total National Health Expenditures',
                        'Subcategory 1': sub1,
                        'Subcategory 2': sub2,
                        'Subcategory 3': sub3_value,
                        'Value': row[col],
                        'Year': row['Year']
                    })
    
    result_df = pd.DataFrame(result_rows)
    return result_df
```

Conjunto de datos final para dashboard:

```python
df_transformado = transform_to_hierarchy(df_analisis)
print(df_transformado.head(100))
print(f"\nTotal de filas: {len(df_transformado)}")
```
