Se puede ver que el archivo es transpuesto. Es decir, las variables están ubicadas en la primera posición de cada fila y le siguen sus respectivos datos. Por lo que se realiza su respectiva transformación y se obtiene:

```python
df = df_raw.T
df.columns = df.iloc[0]   #Se convierte la primera fila en encabezados
df = df[1:]
df.head()
```

```python
df.shape
```

En total, se tiene un conjunto de datos de 545 registros y 64 variables.

```python
df.dtypes
```

Como muestra el resultado, la mayoría de columnas están definidas como object o como float64, por lo que será necesaria una transformación para el tratamiento y su análisis durante la limpieza.

```python
# df.isna().sum()
int((df.isna().any(axis=1)).sum())
```

Asimismo, se observa que en las 64 variables hay presencia de valores faltantes.

Dado que hay números definidos con "," con miles y "-" para NaN, se realiza la siguiente transformación para que afecte el procesamiento.

```python
df = df.replace('"', '', regex=True)
df = df.replace('\.', '', regex=True)
df = df.replace(',', '', regex=True)  
df = df.replace('-', None)          
```
